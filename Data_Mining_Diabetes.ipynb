#Evaluate Models:


```python
models = [
    KNeighborsClassifier(n_neighbors=3),
    SVC(kernel='linear'),
    XGBClassifier(),
    AdaBoostClassifier(),
    RandomForestClassifier(),
    DecisionTreeClassifier(),
    MLPClassifier()
]
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
import pandas as pd

# Create an empty DataFrame to store results
results_df = pd.DataFrame(columns=['Model', 'Train Accuracy', 'Test Accuracy', 'MSE', 'RMSE', 'MAE'])

# Loop through models
for model in models:
    # Fit the model
    model.fit(X_train, y_train)

    # Make predictions on train and test sets
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calculate metrics for train set
    train_accuracy = accuracy_score(y_train, y_train_pred)

    # Calculate metrics for test set
    test_accuracy = accuracy_score(y_test, y_test_pred)
    mse = mean_squared_error(y_test, y_test_pred)
    rmse = mean_squared_error(y_test, y_test_pred, squared=False)  # Calculate RMSE
    mae = mean_absolute_error(y_test, y_test_pred)

    # Append results to DataFrame
    results_df = results_df.append({
        'Model': model.__class__.__name__,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae
    }, ignore_index=True)

# Display the results DataFrame
results_df
```

```python
from sklearn.model_selection import learning_curve
import numpy as np

from sklearn.model_selection import cross_val_score, learning_curve
import numpy as np
import matplotlib.pyplot as plt


for model in models:
    # Use learning_curve to get training and test scores
    train_sizes, train_scores, test_scores = learning_curve(model, X_train, y_train, cv=5, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10))

    # Calculate mean and standard deviation of scores
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    # Plot learning curve
    plt.figure(figsize=(8, 6))
    plt.plot(train_sizes, train_mean, label='Training Accuracy', marker='o')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.15)
    plt.plot(train_sizes, test_mean, label='Testing Accuracy', marker='o')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.15)
    plt.title(f'Learning Curve for {model.__class__.__name__}')
    plt.xlabel('Training Examples')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from mlxtend.classifier import StackingClassifier


knn_model = KNeighborsClassifier()
rf_model = RandomForestClassifier()
dt_model = DecisionTreeClassifier()
xgb_model = XGBClassifier()

meta_model = rf_classifier = RandomForestClassifier(
    min_samples_leaf=4,
    min_samples_split=10,
    n_estimators=100
)
base_models = [knn_model, rf_model, dt_model, xgb_model]

stacking_model = StackingClassifier(classifiers=base_models, meta_classifier=meta_model)

for model in base_models:
    model.fit(X_train, y_train)

stacking_model.fit(X_train, y_train)

base_predictions = [model.predict(X_test) for model in base_models]
stacking_predictions = stacking_model.predict(X_test)

for i, model in enumerate(base_models):
    acc = accuracy_score(y_test, base_predictions[i])
    print(f"Accuracy for {model.__class__.__name__}: {acc}")

acc_stacking = accuracy_score(y_test, stacking_predictions)
print(f"Accuracy for Stacking Model: {acc_stacking}")
```
